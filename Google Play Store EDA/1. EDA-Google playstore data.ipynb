```python
# 1MB = 1024KB
import numpy as np
import pandas as pd

def size_process(item):
    item = str(item).strip()

    # Handle missing or special cases
    if item.lower() in ["varies with device", "nan", "none"]:
        return np.nan

    # Case 1: value in MB → convert to KB
    if item.endswith("M"):
        try:
            return float(item.replace("M", "")) * 1024
        except:
            return np.nan

    # Case 2: value in KB → keep as KB
    elif item.endswith("k") or item.endswith("K"):
        try:
            return float(item.replace("k", "").replace("K", ""))
        except:
            return np.nan

    # Case 3: plain number → assume already KB
    elif item.replace(".", "", 1).isdigit():
        try:
            return float(item)
        except:
            return np.nan

    # Otherwise
    return np.nan




# Test
print(size_process("19M"))   # 19456.0 KB
print(size_process("950k"))  # 950.0 KB
print(size_process("120"))   # 120.0 KB
print(size_process("Varies with device"))  # nan
```

```python
def clean_android_ver(ver):
    if pd.isna(ver) or ver == "":
        return np.nan
    
    ver = str(ver).strip()

    # Handle ranges → take the minimum version
    if "-" in ver:
        ver = ver.split("-")[0].strip()

    # Remove "W" (Android Wear)
    ver = ver.replace("W", "")

    # Keep only first two version parts (major.minor)
    parts = ver.split(".")
    if len(parts) >= 2:
        ver = parts[0] + "." + parts[1]   # e.g. "4.0.3" → "4.0"
    else:
        ver = parts[0]                    # e.g. "7" → "7"

    return ver
```

```python
installs_by_cat = df_copy.groupby("Category")["Installs"].sum().sort_values(ascending=False)
```

```python
print("Most Installed Category:", installs_by_cat.index[0])
```

```python

plt.figure(figsize=(10,6))
sns.barplot(x=installs_by_cat.index, y=installs_by_cat.values, palette="pastel")
plt.title("Total Installs by Category")
plt.ylabel("Total Installs")
plt.xticks(rotation=90)
plt.show()
```

```python
top_installs = (
    df_copy.groupby(["Type", "Category"])["Installs"]
    .sum()
    .reset_index()
    .sort_values(["Type", "Installs"], ascending=[True, False])
)
```

```python
top5_each_type = top_installs.groupby("Type").head(5)
top5_each_type
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=top5_each_type, 
            x="Category", 
            y="Installs", 
            hue="Type", 
            palette="pastel")

plt.title("Top 5 Installed Categories by App Type")
plt.ylabel("Total Installs")
plt.xlabel("Category")
plt.xticks(rotation=45)
plt.legend(title="App Type")
plt.show()
```

```python
most_reviews = (
    df_copy.groupby(["Category", "App"])["Reviews"]
    .sum()
    .reset_index()
    .sort_values("Reviews", ascending=False)
    .head(10)
)
most_reviews
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=most_reviews, 
            x="Reviews", 
            y="App", 
            hue="Category", 
            dodge=False,   # keep one bar per app
            palette="pastel")

plt.title("Top 10 Most Reviewed Apps")
plt.xlabel("Total Reviews")
plt.ylabel("App")
plt.legend(title="Category")
plt.show()
```
