```python
num_cols = ['age', 'traveltime', 'studytime', 'failures','Medu','Fedu','famrel','freetime','goout','Dalc','health','Walc', 'absences', 'G1', 'G2', 'G3']
plt.figure(figsize=(16,6)) 
sns.boxplot(data = df[num_cols], palette = 'Set2')
```

```python
# Age: Mostly 15–18, few older outliers.
# Travel time: Mostly ≤2 units, few long travel outliers.
# Study time: Majority at lower study time.
# Failures: Most have 0, some 1–3 outliers.
# Medu & Fedu: Skewed towards lower parental education.
# Family relation (famrel): Mostly strong (4–5).
# Freetime: Centered around 3–4.
# Goout: Centered around 3, few high.
# Dalc: Daily alcohol mostly very low (1).
# Health: Wide spread, mostly 3–5.
# Walc: Weekend alcohol mostly very low.
# Absences: Many at 0, heavy right-skew with big outliers.
# G1, G2, G3 (grades): Cluster around 10–14, some very low outliers.
```
