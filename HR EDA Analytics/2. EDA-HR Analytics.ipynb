### HR Analytics - Employees Engagement
In this EDA project we will performing HR Analytics - Employees Engagement Analysis which is present in kaggle platform.In this EDA Project we will analyze and visualize our dataset.

The job role of a HR is not that easy as it seems from the outside

The HR's have to actively participate in the recruitment process, helping employees with their issues, maintaining positive work environment, analysing the performance and efficiency, and many more

Among all the job responsibilites of an HR, evaluating the performance and efficiency of the employees is considered the most difficult task

The difficulty level of this task is directly proportional to the no. of employees who work under that particular HR

Thus, to deal with this, we have come with an Exploratory Data Analysis (EDA) project

Here, we'll be performing different analysis and visualizations using the Employees Enagagment Dataset to obtain some valuable insights

```python
#what is the avg enagement score for employees in each department

# Group by Department and calculate average engagement score
avg_engagement = (
    df.groupby("Department")["EngagementSurvey"]
    .mean()
    .reset_index()
    .sort_values("EngagementSurvey", ascending=False)
)

print(avg_engagement)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=avg_engagement, 
            x="Department", 
            y="EngagementSurvey", 
            palette="pastel")

plt.title("Average Engagement Score by Department")
plt.ylabel("Average Engagement Score")
plt.xlabel("Department")
plt.xticks(rotation=45)
plt.show()
```

```python
#internal h/w

#How many employees have been terminated for each position
terminated_count = (
    df[df["Termd"] == 1]
    .groupby("Position")["Employee_Name"]
    .count()
    .reset_index()
    .rename(columns={"Employee_Name": "Terminated_Employees"})
)

print(terminated_count)
```

```python
plt.figure(figsize=(25,25))
plt.pie(terminated_count["Terminated_Employees"], 
        labels=terminated_count["Position"], 
        autopct="%1.1f%%", 
        colors=sns.color_palette("pastel"),
       textprops={"fontsize":12})

plt.title("Proportion of Terminations by Position")
plt.show()
```

```python
df.columns
```

```python
#how many employees have been terminated for each reason
df[df['Termd'] == 1].groupby('TermReason')['Employee_Name'].count()
terminated = df[df['Termd'] == 1]
terminated_reason_count = (
    terminated.groupby("TermReason")["Employee_Name"]
    .count()
    .reset_index()
    .rename(columns={"Employee_Name": "Terminated_Employees"})
)
```

```python
plt.figure(figsize=(10,6))
ax = sns.barplot(
    data=terminated_reason_count,
    y="TermReason",
    x="Terminated_Employees",
    palette="pastel"
)

# Show counts on bars
for index in range(len(terminated_reason_count)):
    value = terminated_reason_count["Terminated_Employees"].iloc[index]
    reason = terminated_reason_count["TermReason"].iloc[index]

    # Write the value on the bar
    ax.text(value + 0.5, index, str(value), color="black", va="center", fontsize=10)

plt.title("Number of Terminated Employees by Reason", fontsize=14)
plt.xlabel("Terminated Employees")
plt.ylabel("Termination Reason")
plt.show()
```

```python
dept_stats = df.groupby("Department").agg({
    "Absences": "sum",
    "EngagementSurvey": "mean"
}).reset_index()
dept_stats
```

```python
plt.figure(figsize=(14,6))

# Plot total absences
plt.subplot(1,2,1)
sns.barplot(data=dept_stats, 
            x="Department", 
            y="Absences", 
            palette="pastel")
plt.title("Total Absences by Department")
plt.xticks(rotation=45)
plt.ylabel("Total Absences")

# Plot average engagement survey
plt.subplot(1,2,2)
sns.barplot(data=dept_stats, 
            x="Department", 
            y="EngagementSurvey", 
            palette="muted")
plt.title("Average Engagement Survey by Department")
plt.xticks(rotation=45)
plt.ylabel("Average Engagement Score")

plt.tight_layout()
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(10,6))

# Bar plot for Absences
sns.barplot(data=dept_stats, x="Department", y="Absences", ax=ax1, color="skyblue")
ax1.set_ylabel("Total Absences", color="blue")
ax1.set_title("Absences vs Engagement Survey by Department")

# Line plot for Engagement Survey on second axis
ax2 = ax1.twinx()
sns.lineplot(data=dept_stats, x="Department", y="EngagementSurvey", ax=ax2, color="red", marker="o")
ax2.set_ylabel("Average Engagement Survey", color="red")

plt.xticks(rotation=45)
plt.show()
```

```python
#What is maximum salary and minimum days late in the last 30 days for employees in each position?
#What is earliest and latest 
```

```python
# Group by Position and calculate max salary + min days late
pos_stats = df.groupby("Position").agg({
    "Salary": "max",
    "DaysLateLast30": "min"
}).reset_index()

pos_stats
```

```python
plt.figure(figsize=(14,5))

# Max Salary per position
plt.subplot(1,2,1)
plt.bar(pos_stats["Position"], pos_stats["Salary"], color="skyblue")
plt.title("Max Salary by Position")
plt.xticks(rotation=90)
plt.ylabel("Salary")

# Min Days Late per position
plt.subplot(1,2,2)
plt.bar(pos_stats["Position"], pos_stats["DaysLateLast30"], color="salmon")
plt.title("Min Days Late in Last 30 Days by Position")
plt.xticks(rotation=90)
plt.ylabel("Days Late")
```
